# 1. País com a maior popularidade

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
df = pd.read_csv("social-media.csv", encoding='latin1')

# Find the most popular country
most_popular_country = df["Country"].value_counts().idxmax()

# Print the result
print(f"O país mais popular é o: {most_popular_country}")

# Create a bar chart to visualize the results
plt.figure(figsize=(10, 6))
df["Country"].value_counts().plot(kind="bar")
plt.title("Popularity of Countries")
plt.xlabel("Country")
plt.ylabel("Count")
plt.show()
```

Como é possivel observar nos resultados gerados o país com o maior nível de popularidade é a Turquia

# 2. Interação dos usuários acima de 30 anos

```python
# Calculate the total likes for users under 30 and over 30
total_likes_under_30 = df[df["Age"] < 30]["TotalLikes"].sum()
total_likes_over_30 = df[df["Age"] >= 30]["TotalLikes"].sum()

total_likes_under_30 = df[df["Age"] < 30]["TotalLikes"].sum()
print(f"Proporção de Likes por Faixa Etária acima de 30 anos: {total_likes_under_30}")

# Create a list for the labels and sizes
labels = ['Abaixo de 30', '30 ou mais']
sizes = [total_likes_under_30, total_likes_over_30]



# Create a pie chart
plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.title("Proporção de Likes por Faixa Etária")
plt.show()
```

Como é possivel observar o grau de interação como o esperado é menor em relação a usuários mais jovens

# 3. correlação entre idade e duração de uso

```python
# Calculate the correlation coefficient
correlation = df["Age"].corr(df["UsageDuraiton"])
# Print the result
print(f"correlação entre idade e duração de uso: {correlation}")



# Create a scatter plot to visualize the relationship
plt.figure(figsize=(8, 6))
plt.scatter(df["Age"], df["UsageDuraiton"])
plt.title("correlação entre idade e duração de uso")
plt.xlabel("Age")
plt.ylabel("UsageDuraiton")
plt.grid(True)
plt.show()
```

Com base nesse gráfico, não podemos concluir que existe uma relação direta entre a idade e o tempo de uso. Para ter uma análise mais precisa, seria necessário analisar um conjunto de dados maior e considerar outros fatores que possam influenciar essa relação.

# 4. média de curtidas por país

```python
# Calculate the average likes by country
average_likes_by_country = df.groupby("Country")["TotalLikes"].mean()

# Print the result
print(average_likes_by_country)

# Create a bar chart to visualize the average likes
plt.figure(figsize=(10, 6))
average_likes_by_country.plot(kind="bar")
plt.title("média de curtidas por país")
plt.xlabel("País")
plt.ylabel("média de curtidas")
plt.show()
```

De acordo com as informações é possível indentificar que o país com a maior média de curtidas é Singapura

# 5. Proporção de Usuários por Nível de Utilização

```python
# Count the number of users with usage duration greater than 5
num_users_high_usage = (df["UsageDuraiton"] > 5).sum()
num_users_low_usage = len(df) - num_users_high_usage

num_users_high_usage = (df["UsageDuraiton"] > 5).sum()
print(f"Proporção de Usuários por Nível de Utilização maior que 5 horas: {num_users_high_usage}")

# Create a list for the labels and sizes
labels = ['Alta Utilização', 'Baixa Utilização']
sizes = [num_users_high_usage, num_users_low_usage]

# Create a pie chart
plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.title("Proporção de Usuários por Nível de Utilização")
plt.show()
```

A proporção é baixa tendo em vista que só 23.8% dos usuários usa uma média maior que 5 horas por dia

# 6. Média de uso por Faixa Etária

```python
# Calculando a média de likes por faixa etária
media_likes_por_faixa = df.groupby('Age')['UsageDuraiton'].mean()

# Criando o gráfico de barras
plt.figure(figsize=(10, 6))
media_likes_por_faixa.plot(kind='bar')
plt.title('Tempo de tela por Faixa Etária')
plt.xlabel('Faixa Etária')
plt.ylabel('Uso de duração')
plt.xticks(rotation=45)  # Rotaciona os rótulos do eixo x para melhor visualização
plt.grid(axis='y')
plt.show()
```

A média de uso apresenta um pico na idade de 25 anos tendo uma queda conforme a idade vai sofrendo avanços achando o seu menor ponto aos 60 anos
